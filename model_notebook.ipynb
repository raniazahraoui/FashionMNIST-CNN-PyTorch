```python
# Fashion-MNIST Classification avec CNN et PyTorch
# Notebook Colab pour l'apprentissage complet du Deep Learning

"""
Ce notebook implémente un réseau de neurones convolutionnel (CNN) pour classifier
les images du dataset Fashion-MNIST (vêtements et accessoires).

Objectifs pédagogiques:
- Comprendre l'architecture CNN
- Maîtriser PyTorch pour le Deep Learning
- Apprendre le preprocessing des données
- Visualiser les résultats d'entraînement
- Sauvegarder et charger un modèle
"""

# ============================================================================
# 1. IMPORTATION DES BIBLIOTHÈQUES
# ============================================================================

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from tqdm import tqdm
import os

print(f"Version PyTorch: {torch.__version__}")
print(f"CUDA disponible: {torch.cuda.is_available()}")

# ============================================================================
# 2. CONFIGURATION ET HYPERPARAMÈTRES
# ============================================================================

# Configuration du device (GPU si disponible, sinon CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device utilisé: {device}")

# Hyperparamètres du modèle et de l'entraînement
BATCH_SIZE = 64          # Nombre d'images par batch
LEARNING_RATE = 0.001    # Taux d'apprentissage
NUM_EPOCHS = 10          # Nombre d'époques d'entraînement
INPUT_SIZE = 28          # Taille des images (28x28 pixels)
NUM_CLASSES = 10         # 10 catégories de vêtements

# Classes du dataset Fashion-MNIST
class_names = [
    'T-shirt/top',    # 0
    'Trouser',        # 1
    'Pullover',       # 2
    'Dress',          # 3
    'Coat',           # 4
    'Sandal',         # 5
    'Shirt',          # 6
    'Sneaker',        # 7
    'Bag',            # 8
    'Ankle boot'      # 9
]

# ============================================================================
# 3. PRÉPARATION DES DONNÉES
# ============================================================================

"""
Transformations appliquées aux images:
- ToTensor(): Convertit les images PIL en tenseurs PyTorch (0-1)
- Normalize(): Normalise avec moyenne=0.5 et écart-type=0.5
  Cela centre les données entre -1 et 1 pour faciliter l'apprentissage
"""

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # Mean=0.5, Std=0.5
])

# Téléchargement et chargement du dataset
print("Téléchargement du dataset Fashion-MNIST...")

train_dataset = datasets.FashionMNIST(
    root='./data',           # Dossier de sauvegarde
    train=True,              # Dataset d'entraînement
    download=True,           # Télécharger si non présent
    transform=transform
)

test_dataset = datasets.FashionMNIST(
    root='./data',
    train=False,             # Dataset de test
    download=True,
    transform=transform
)

# DataLoaders pour charger les données par batches
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,            # Mélange les données à chaque époque
    num_workers=2            # Processus parallèles pour le chargement
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2
)

print(f"Dataset d'entraînement: {len(train_dataset)} images")
print(f"Dataset de test: {len(test_dataset)} images")
print(f"Nombre de batches d'entraînement: {len(train_loader)}")

# ============================================================================
# 4. VISUALISATION DES DONNÉES
# ============================================================================

def visualize_samples(dataset, num_samples=10):
    """
    Affiche un échantillon d'images du dataset

    Args:
        dataset: Dataset PyTorch
        num_samples: Nombre d'images à afficher
    """
    fig, axes = plt.subplots(2, 5, figsize=(12, 5))
    axes = axes.ravel()

    for i in range(num_samples):
        image, label = dataset[i]
        # Dénormalisation pour affichage (de [-1,1] vers [0,1])
        image = image.squeeze() * 0.5 + 0.5

        axes[i].imshow(image, cmap='gray')
        axes[i].set_title(f"{class_names[label]}")
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

print("\nExemples d'images du dataset:")
visualize_samples(train_dataset)

# ============================================================================
# 5. DÉFINITION DU MODÈLE CNN
# ============================================================================

class FashionCNN(nn.Module):
    """
    Réseau de Neurones Convolutionnel pour Fashion-MNIST

    Architecture:
    - 2 blocs convolutionnels (Conv + ReLU + MaxPool)
    - 2 couches fully-connected
    - Dropout pour la régularisation
    """

    def __init__(self):
        super(FashionCNN, self).__init__()

        # ===== BLOC CONVOLUTIONNEL 1 =====
        # Conv1: 1 canal (grayscale) -> 32 filtres, kernel 3x3
        self.conv1 = nn.Conv2d(
            in_channels=1,      # Images en niveaux de gris
            out_channels=32,    # 32 filtres de convolution
            kernel_size=3,      # Fenêtre 3x3
            padding=1           # Padding pour garder la taille
        )
        # Sortie: 32 x 28 x 28

        # ReLU: Fonction d'activation (introduit la non-linéarité)
        self.relu1 = nn.ReLU()

        # MaxPool: Réduit les dimensions spatiales par 2
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        # Sortie: 32 x 14 x 14

        # ===== BLOC CONVOLUTIONNEL 2 =====
        # Conv2: 32 canaux -> 64 filtres
        self.conv2 = nn.Conv2d(
            in_channels=32,
            out_channels=64,
            kernel_size=3,
            padding=1
        )
        # Sortie: 64 x 14 x 14

        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        # Sortie: 64 x 7 x 7

        # ===== COUCHES FULLY-CONNECTED =====
        # Aplatissement: 64 x 7 x 7 = 3136 neurones
        self.flatten = nn.Flatten()

        # FC1: 3136 -> 128 neurones
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.relu3 = nn.ReLU()

        # Dropout: Désactive aléatoirement 50% des neurones (régularisation)
        self.dropout = nn.Dropout(0.5)

        # FC2: 128 -> 10 classes (sortie finale)
        self.fc2 = nn.Linear(128, NUM_CLASSES)

    def forward(self, x):
        """
        Propagation avant (forward pass)

        Args:
            x: Batch d'images [batch_size, 1, 28, 28]

        Returns:
            Logits pour chaque classe [batch_size, 10]
        """
        # Bloc convolutionnel 1
        x = self.conv1(x)      # [batch, 32, 28, 28]
        x = self.relu1(x)
        x = self.pool1(x)      # [batch, 32, 14, 14]

        # Bloc convolutionnel 2
        x = self.conv2(x)      # [batch, 64, 14, 14]
        x = self.relu2(x)
        x = self.pool2(x)      # [batch, 64, 7, 7]

        # Couches fully-connected
        x = self.flatten(x)    # [batch, 3136]
        x = self.fc1(x)        # [batch, 128]
        x = self.relu3(x)
        x = self.dropout(x)    # Régularisation
        x = self.fc2(x)        # [batch, 10]

        return x

# Instanciation du modèle
model = FashionCNN().to(device)

# Affichage de l'architecture
print("\n" + "="*70)
print("ARCHITECTURE DU MODÈLE")
print("="*70)
print(model)

# Calcul du nombre de paramètres
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nNombre total de paramètres: {total_params:,}")
print(f"Paramètres entraînables: {trainable_params:,}")

# ============================================================================
# 6. FONCTION DE PERTE ET OPTIMISEUR
# ============================================================================

"""
CrossEntropyLoss: Fonction de perte pour la classification multi-classes
Combine LogSoftmax + NLLLoss en une seule fonction

Adam: Optimiseur adaptatif avec moment
- Ajuste automatiquement le taux d'apprentissage
- Fonctionne bien pour la plupart des problèmes
"""

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# ============================================================================
# 7. FONCTIONS D'ENTRAÎNEMENT ET D'ÉVALUATION
# ============================================================================

def train_epoch(model, loader, criterion, optimizer, device):
    """
    Entraîne le modèle pendant une époque

    Returns:
        Perte moyenne et précision sur l'époque
    """
    model.train()  # Mode entraînement (active dropout, etc.)

    running_loss = 0.0
    correct = 0
    total = 0

    # Barre de progression
    pbar = tqdm(loader, desc='Entraînement')

    for images, labels in pbar:
        # Transfert sur GPU si disponible
        images, labels = images.to(device), labels.to(device)

        # 1. Reset des gradients
        optimizer.zero_grad()

        # 2. Forward pass
        outputs = model(images)

        # 3. Calcul de la perte
        loss = criterion(outputs, labels)

        # 4. Backward pass (calcul des gradients)
        loss.backward()

        # 5. Mise à jour des poids
        optimizer.step()

        # Statistiques
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # Mise à jour de la barre de progression
        pbar.set_postfix({
            'loss': f'{loss.item():.4f}',
            'acc': f'{100 * correct / total:.2f}%'
        })

    epoch_loss = running_loss / len(loader)
    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc


def evaluate(model, loader, criterion, device):
    """
    Évalue le modèle sur un dataset

    Returns:
        Perte moyenne et précision
    """
    model.eval()  # Mode évaluation (désactive dropout, etc.)

    running_loss = 0.0
    correct = 0
    total = 0

    # Désactivation du calcul des gradients (économie de mémoire)
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Évaluation'):
            images, labels = images.to(device), labels.to(device)

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Statistiques
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(loader)
    epoch_acc = 100 * correct / total

    return epoch_loss, epoch_acc

# ============================================================================
# 8. BOUCLE D'ENTRAÎNEMENT PRINCIPALE
# ============================================================================

print("\n" + "="*70)
print("DÉBUT DE L'ENTRAÎNEMENT")
print("="*70)

# Historique pour la visualisation
history = {
    'train_loss': [],
    'train_acc': [],
    'test_loss': [],
    'test_acc': []
}

best_acc = 0.0

for epoch in range(NUM_EPOCHS):
    print(f"\nÉpoque [{epoch+1}/{NUM_EPOCHS}]")
    print("-" * 70)

    # Entraînement
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)

    # Évaluation
    test_loss, test_acc = evaluate(model, test_loader, criterion, device)

    # Sauvegarde de l'historique
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['test_loss'].append(test_loss)
    history['test_acc'].append(test_acc)

    # Affichage des résultats
    print(f"\nRésultats:")
    print(f"  Train - Loss: {train_loss:.4f}, Acc: {train_acc:.2f}%")
    print(f"  Test  - Loss: {test_loss:.4f}, Acc: {test_acc:.2f}%")

    # Sauvegarde du meilleur modèle
    if test_acc > best_acc:
        best_acc = test_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'test_acc': test_acc,
        }, 'best_model.pth')
        print(f"  ✓ Meilleur modèle sauvegardé! (Acc: {best_acc:.2f}%)")

print("\n" + "="*70)
print("ENTRAÎNEMENT TERMINÉ")
print("="*70)
print(f"Meilleure précision sur le test: {best_acc:.2f}%")

# ============================================================================
# 9. VISUALISATION DES RÉSULTATS
# ============================================================================

def plot_training_history(history):
    """Affiche les courbes de perte et précision"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    epochs = range(1, len(history['train_loss']) + 1)

    # Courbe de perte
    ax1.plot(epochs, history['train_loss'], 'b-', label='Train', linewidth=2)
    ax1.plot(epochs, history['test_loss'], 'r-', label='Test', linewidth=2)
    ax1.set_title('Évolution de la Perte', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Époque')
    ax1.set_ylabel('Perte')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Courbe de précision
    ax2.plot(epochs, history['train_acc'], 'b-', label='Train', linewidth=2)
    ax2.plot(epochs, history['test_acc'], 'r-', label='Test', linewidth=2)
    ax2.set_title('Évolution de la Précision', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Époque')
    ax2.set_ylabel('Précision (%)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

plot_training_history(history)

# ============================================================================
# 10. MATRICE DE CONFUSION
# ============================================================================

def compute_confusion_matrix(model, loader, device):
    """Calcule la matrice de confusion"""
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.numpy())

    return confusion_matrix(all_labels, all_preds)

# Calcul et affichage
cm = compute_confusion_matrix(model, test_loader, device)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matrice de Confusion', fontsize=16, fontweight='bold')
plt.ylabel('Vraie Classe')
plt.xlabel('Classe Prédite')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Rapport de classification
print("\nRapport de Classification:")
print("="*70)
# Récupération des prédictions pour le rapport
model.eval()
all_preds = []
all_labels = []
with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

print(classification_report(all_labels, all_preds, target_names=class_names))

# ============================================================================
# 11. VISUALISATION DES PRÉDICTIONS
# ============================================================================

def visualize_predictions(model, dataset, device, num_samples=10):
    """Affiche des prédictions du modèle"""
    model.eval()

    fig, axes = plt.subplots(2, 5, figsize=(14, 6))
    axes = axes.ravel()

    indices = np.random.choice(len(dataset), num_samples, replace=False)

    with torch.no_grad():
        for i, idx in enumerate(indices):
            image, true_label = dataset[idx]
            image_display = image.squeeze() * 0.5 + 0.5  # Dénormalisation

            # Prédiction
            image_input = image.unsqueeze(0).to(device)
            output = model(image_input)
            _, predicted = torch.max(output, 1)
            pred_label = predicted.item()

            # Probabilités
            probs = torch.nn.functional.softmax(output, dim=1)
            confidence = probs[0][pred_label].item() * 100

            # Affichage
            axes[i].imshow(image_display, cmap='gray')
            color = 'green' if pred_label == true_label else 'red'
            axes[i].set_title(
                f"Vrai: {class_names[true_label]}\n"
                f"Prédit: {class_names[pred_label]}\n"
                f"Confiance: {confidence:.1f}%",
                color=color, fontsize=9
            )
            axes[i].axis('off')

    plt.tight_layout()
    plt.show()

print("\nExemples de prédictions:")
visualize_predictions(model, test_dataset, device)

# ============================================================================
# 12. SAUVEGARDE FINALE DU MODÈLE
# ============================================================================

# Chemin de sauvegarde
model_path = 'fashion_mnist_cnn_final.pth'

# Sauvegarde complète
torch.save({
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'history': history,
    'best_acc': best_acc,
    'class_names': class_names,
    'hyperparameters': {
        'batch_size': BATCH_SIZE,
        'learning_rate': LEARNING_RATE,
        'num_epochs': NUM_EPOCHS
    }
}, model_path)

print(f"\n✓ Modèle sauvegardé: {model_path}")

# Pour Google Colab: téléchargement du modèle
try:
    from google.colab import files
    files.download(model_path)
    print(f"✓ Téléchargement du modèle initié!")
except:
    print(f"✓ Modèle disponible localement: {model_path}")

# ============================================================================
# 13. CODE POUR CHARGER LE MODÈLE EN LOCAL
# ============================================================================

print("\n" + "="*70)
print("CODE POUR UTILISER LE MODÈLE EN LOCAL")
print("="*70)

code_local = """
# ===== Chargement du modèle en local =====

import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image

# 1. Redéfinir l'architecture du modèle
class FashionCNN(nn.Module):
    def __init__(self):
        super(FashionCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, 2)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.relu3 = nn.ReLU()
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = self.flatten(x)
        x = self.dropout(self.relu3(self.fc1(x)))
        return self.fc2(x)

# 2. Charger le modèle
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = FashionCNN().to(device)

checkpoint = torch.load('fashion_mnist_cnn_final.pth', map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()

# 3. Prédire sur une nouvelle image
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

def predict_image(image_path):
    # Preprocessing
    transform = transforms.Compose([
        transforms.Grayscale(),
        transforms.Resize((28, 28)),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])

    image = Image.open(image_path)
    image = transform(image).unsqueeze(0).to(device)

    # Prédiction
    with torch.no_grad():
        output = model(image)
        probs = torch.nn.functional.softmax(output, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        confidence = probs[0][pred].item() * 100

    return class_names[pred], confidence

# Exemple d'utilisation
# prediction, confidence = predict_image('mon_image.png')
# print(f"Prédiction: {prediction} ({confidence:.2f}% de confiance)")
"""

print(code_local)

print("\n" + "="*70)
print("NOTEBOOK TERMINÉ AVEC SUCCÈS!")
print("="*70)
print(f"✓ Modèle entraîné avec {best_acc:.2f}% de précision")
print(f"✓ Fichier sauvegardé: {model_path}")
print("✓ Prêt pour utilisation en local")
print("="*70)
```
